```python
import tensorflow as tf
```

```python
#행렬을 이용하여 Data표시
x_data = [[73.,80.,75. ],[93.,88.,93.],[89.,90.,90.],[96.,98.,100.],[73.,66.,70.] ]
y_data = [[152.],[185.],[180.],[196.],[170.]]
```

```python
# shpae = [None] => 몇개의 데이터 수가 들어갈지 모름,  3은 하나의 data에 대한 feature의 수 
X = tf.placeholder(tf.float32 , shape=[None, 3])
Y = tf.placeholder(tf.float32 , shape=[None,1])
```

```python
W = tf.Variable(tf.random_normal([3,1]) , name = "weight")
b = tf.Variable(tf.random_normal([1]) , name = "bias")
```

```python
# matmul => matrix multiply 메트릭스의 곱
hypothesis = tf.matmul(X,W) + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
opitmizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
```

```python
train = opitmizer.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
for step in range(10000) :
    cost_val , hy_val ,_ = sess.run([cost , hypothesis , train] , feed_dict ={ X : x_data, Y : y_data})
    if step % 100 == 0 :
        print( step , "\ncost : " , cost_val , "\nPrediction :" , hy_val )
```

```python
print(sess.run(hypothesis , feed_dict={ X : [[73,80,75]]}))
```
