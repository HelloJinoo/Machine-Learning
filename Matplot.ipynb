```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
W = tf.placeholder(tf.float32)
```

```python
W = tf.Variable(tf.random_normal([1]) , name="weight")
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
hypothesis = X * W 
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01) 
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
W_val =[]
cost_val = []
```

```python
for i in range(-30, 50) :
    feed_W = i * 0.1
    curr_cost , curr_W = sess.run([cost,W] , feed_dict={ W : feed_W , Y : [1,2,3] , X : [1,2,3]})
    
    #구한 cost와 W를 각각의 리스트에 append
    W_val.append(curr_W)   
    cost_val.append(curr_cost)
```

```python
# list에 추가된 데이터를 가지고 그래프를 생성
plt.plot(W_val , cost_val)
plt.show()
```

```python
learning_rate = 0.1
gradient = tf.reduce_mean(tf.square(hypothesis - Y))
descent = W - learning_rate * gradient
update = W.assign(descent)
```

```python
for step in range(1000) :
    sess.run( update, feed_dict = { X : [1, 2, 3] , Y : [1, 2 ,3]} )
    print(step , sess.run(cost , feed_dict= {X : [1,2,3] , Y : [1,2,3]})  , sess.run(W) )
```
