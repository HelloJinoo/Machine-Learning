```python
import os
import pandas as pd
import json
from pandas import DataFrame
```

```python
directory = os.listdir('sokulee/')
all_user_step = []
all_user_heart = []
all_user_sleep = []
```

```python
for i in range(0,len(directory)) :
    file = os.listdir('sokulee/'+directory[i])
    for j in file : 
        total_data = json.load(open('sokulee/'+ directory[i] + '/'+ j))
        if 'heart' in j :
            if len(total_data["activities-heart-intraday"]['dataset']) > 0 :
                need_data = total_data["activities-heart-intraday"]['dataset']
                df = DataFrame(need_data)
                df['DATE'] = total_data["activities-heart"][0]['dateTime']
                df['USERNAME'] = directory[i]
                all_user_heart.append(df)
        elif 'sleep' in j : 
            if len(total_data['sleep']) > 0 :
                need_data = total_data['sleep'][0]
                df = DataFrame()
                df['USERNAME'] = [directory[i]]
                df['DATE'] = [need_data["dateOfSleep"]]
                df['STARTTIME'] = [need_data['startTime'].split('T')[1]]
                df['MINUTESASLEEP'] = [need_data['minutesAsleep']]
                df['MINUTESAWAKE'] = [need_data['minutesAwake']]
                df['TIMEINBED'] = [need_data['timeInBed']]
                all_user_sleep.append(df)
        else :            
            if 'activities-steps-intraday' in total_data.keys() : 
                need_data = total_data['activities-steps-intraday']['dataset']
                df = DataFrame(need_data)
                df['DATE'] = total_data['activities-steps'][0]["dateTime"]
                df['USERNAME'] = directory[i]
                df['HOUR'] = df['time'].apply(lambda x : int(x.split(":")[0]))
                df['MINUTE'] = df['time'].apply(lambda x : int(x.split(":")[1]))
                all_user_step.append(df)
            
        
        
    
```

```python
df_all_user_heart = pd.concat(all_user_heart).reset_index()
df_all_user_step = pd.concat(all_user_step).reset_index()
df_all_user_sleep = pd.concat(all_user_sleep).reset_index()
```

```python
del df_all_user_heart['index']
del df_all_user_step['index']
del df_all_user_sleep['index']
df_all_user_step = df_all_user_step.rename(index=str, columns={"time": "TIME", "value": "STEPS"})
df_all_user_heart = df_all_user_heart.rename(index=str, columns={'time' : 'TIME' , 'value' : 'VALUE'})
df_all_user_heart = df_all_user_heart.sort_values(['DATE', 'USERNAME'])
```

```python
heart_order_col = ['USERNAME' ,'DATE', 'TIME', 'VALUE']
df_all_user_heart = df_all_user_heart[heart_order_col]
steps_order_col = ['USERNAME' ,'DATE', 'TIME', 'HOUR' , 'MINUTE' , 'STEPS']
df_all_user_step = df_all_user_step[steps_order_col]
sleep_order_col = ['USERNAME' ,'DATE', 'STARTTIME', 'MINUTESASLEEP' , 'MINUTESAWAKE' , 'TIMEINBED']
df_all_user_sleep = df_all_user_sleep[sleep_order_col]
```
