```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
all_hearts_df = pd.read_csv('./data/all_user_hearts.csv')
all_sleeps_df = pd.read_csv('./data/all_user_sleeps.csv')
all_steps_df = pd.read_csv('./data/all_user_steps.csv')
```

```python
del all_sleeps_df['Unnamed: 0']
del all_hearts_df['Unnamed: 0']
del all_steps_df['Unnamed: 0']
```

## [1-a-i]평균걸음수분포

```python
step_per_day = all_steps_df.groupby(['USERNAME' , 'DATE'])['STEPS'].sum().reset_index()
```

```python
mean_steps_per_day = step_per_day.groupby(['USERNAME'])['STEPS'].mean()
mean_steps_per_day  = mean_steps_per_day.reset_index()
```

```python
mean_steps_per_day
```

```python
mean_steps_per_day.hist()
plt.xlabel('Steps')
plt.ylabel('# of user')
plt.show()
```

```python
steps_list = [int(sleep)for sleep in mean_steps_per_day['STEPS']]
steps_list_count = np.zeros(max(steps_list)+1)

for step in steps_list :
    steps_list_count[int(step)] += 1
    
step_cumsum = steps_list_count.cumsum()
step_cdf = [cumsum/step_cumsum[-1] for cumsum in step_cumsum]
print('하루 10,000보를 달성하지 못한 경우 : {0}%'.format(round(step_cdf[10000]*100,2)))
plt.figure(figsize = (8,6))
plt.plot(step_cdf)
plt.axvline(x=10000, color='r', linewidth=1 , linestyle='--')
plt.xlabel('Steps')
plt.ylabel('CDF')
plt.show()
```

## [1-a-ii]평균수면량분포

```python
mean_sleep_per_day = all_sleeps_df.groupby(['USERNAME'])['MINUTESASLEEP'].mean().reset_index()
```

```python
mean_sleep_per_day
```

```python
mean_sleep_per_day.hist()
plt.xlabel('Minute a sleep')
plt.ylabel('# of user')
plt.show()
```

```python
import numpy as np 
sleep_list = [int(sleep)for sleep in mean_sleep_per_day['MINUTESASLEEP']]
sleep_list_count = np.zeros(max(sleep_list)+1)

for sleep in sleep_list :
    sleep_list_count[int(sleep)] += 1
    
sleep_cumsum = sleep_list_count.cumsum()
sleep_cdf = [cumsum/sleep_cumsum[-1] for cumsum in sleep_cumsum]
print('하루 수면 시간이 6시간이 안되는 경우 : {0}%'.format(round(sleep_cdf[360]*100,2)))
plt.figure(figsize = (8,6))
plt.plot(sleep_cdf)
plt.xlim([200,450])
plt.axvline(x=360, color='r', linewidth=1 , linestyle='--')
plt.xlabel('Minute a sleep')
plt.ylabel('CDF')
plt.show()
```

## [1-a-iii]평균심박수분포

```python
mean_heart_per_day = all_hearts_df.groupby(['USERNAME'])['VALUE'].mean().reset_index()
```

```python
mean_heart_per_day
```

```python
mean_heart_per_day.hist()
plt.xlabel('Value of hearts')
plt.ylabel('# of user')
plt.show()
```

```python
heart_list = [int(heart) for heart in mean_heart_per_day['VALUE']]
heart_list_count = np.zeros(max(heart_list)+1)

sigma = 25

for heart in heart_list :
    heart_list_count[int(heart)] += 1
    
heart_cumsum = heart_list_count.cumsum()
heart_cdf = [cumsum/heart_cumsum[-1] for cumsum in heart_cumsum]
plt.figure(figsize = (8,6))
plt.plot(heart_cdf, color='b')
print('하루 평균 심박수가 60이하인 사람들 : {0}%'.format(round(heart_cdf[60]*100,2)))
plt.xlabel('Heartrate')
plt.ylabel('CDF')
plt.xlim([50,93])
plt.axvline(x=60, color='r', linewidth=1 , linestyle='--')



plt.plot(heart_cdf)  # matplotlib의 plot 기능을 이용하여 (x, y) 좌표 표수
plt.show()  
```

## [1-b]걸음,수면상관관계요인


```python
weather_df = pd.read_csv('./data/sokulee_weather.csv')
del weather_df['STATION']
weather_df
```

```python
weather_df['DATETIME'] = weather_df['DATETIME'].apply(lambda x : x.split(" ")[0])
weather_df = weather_df.rename(index=str , columns={'DATETIME' : 'DATE'})
weather_df = weather_df.fillna(0)
```

```python
weather_df
```

```python
temper_max_per_day = pd.DataFrame(weather_df.groupby(['DATE'])['TEMPERATURE'].max())
wind_mean_per_day = pd.DataFrame(weather_df.groupby(['DATE'])['WINDSPEED'].mean())
humidity_mean_per_day = pd.DataFrame(weather_df.groupby(['DATE'])['HUMIDITY'].mean())
totalrainfall_per_day = pd.DataFrame(weather_df.groupby(['DATE'])['RAINFALL'].sum())

temper_max_per_day = temper_max_per_day.reset_index()
totalrainfall_per_day = totalrainfall_per_day.reset_index()
wind_mean_per_day = wind_mean_per_day.reset_index()
humidity_mean_per_day = humidity_mean_per_day.reset_index()
```

```python
humidity_mean_per_day
```

```python
merge_weather = pd.merge(temper_max_per_day, totalrainfall_per_day)
merge_weather = pd.merge(merge_weather,wind_mean_per_day )
merge_weather = pd.merge(merge_weather , humidity_mean_per_day)
```

```python
merge_weather
```

```python
steps_per_oneday = (step_per_day.groupby(['DATE'])['STEPS'].mean()).reset_index()
sleep_per_oneday = (all_sleeps_df.groupby(['DATE'])['MINUTESASLEEP'].mean()).reset_index()
```

```python
sleep_per_oneday
```

```python
weather_steps_sleep = pd.merge(merge_weather , steps_per_oneday)
weather_steps_sleep = pd.merge(weather_steps_sleep , sleep_per_oneday )
weather_steps_sleep['WEEKDAY'] = weather_steps_sleep['DATE'].apply(lambda x : pd.to_datetime(x).weekday())
```

```python
weather_steps_sleep
```

```python
df_subset = weather_steps_sleep[['STEPS', 'MINUTESASLEEP', 'TEMPERATURE', 'RAINFALL', 'WINDSPEED', 'HUMIDITY','WEEKDAY']]
axes = pd.plotting.scatter_matrix(df_subset, figsize=(15,20) , alpha = 0.5)

corr = df_subset.corr().as_matrix()

for i, j in zip(*plt.np.triu_indices_from(axes , k=1)) :
    axes[i,j].annotate("%.3f" %corr[i,j], (0.8,0.8) , xycoords = 'axes fraction', ha='center' , va = 'center')
plt.show()   
```

## [1-b-ii]최고기온

```python
fig, ax1 = plt.subplots(figsize=(10,5))
ax1.plot(weather_steps_sleep['DATE'],weather_steps_sleep['TEMPERATURE'] , 'r-' , linewidth=1.5 )
ax1.set_ylabel('TEMPERATURE')
ax1.set_xlabel('DATE')
ax1.set_label('Temperature')
ax1.legend()
ax1.set_xticks(['2016-04-01','2016-04-11','2016-04-21','2016-05-01','2016-05-11'])
ax1.set_xlim(weather_steps_sleep['DATE'][0] ,weather_steps_sleep['DATE'][len(weather_steps_sleep['DATE'])-1] )


ax2 = ax1.twinx()
ax2.bar(weather_steps_sleep['DATE'], weather_steps_sleep['STEPS'] , align='center' ,alpha=0.4 ,label='Step')
ax2.set_ylabel('STEPS')
ax2.set_label('Step')
ax2.legend()
ax2.set_xticks(['2016-04-01','2016-04-11','2016-04-21','2016-05-01','2016-05-11'])
ax2.set_xlim(weather_steps_sleep['DATE'][0] ,weather_steps_sleep['DATE'][len(weather_steps_sleep['DATE'])-1] )

plt.tight_layout()

plt.show()
```
